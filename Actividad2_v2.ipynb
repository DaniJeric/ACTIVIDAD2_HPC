```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.interpolate import barycentric_interpolate, lagrange
import sympy as sp
from time import time

# Definir las funciones
f1 = lambda x: np.sin(x)
f2 = lambda x: np.exp(-20 * x**2)
f3 = lambda x: 1 / (1 + 25 * x**2)

# Generar nodos
def generate_nodes(n, method='equiespaciados', func=None):
    if func == f1:
        interval = [0, 6 * np.pi]
    else:
        interval = [-1, 1]
    
    if method == 'equiespaciados':
        return np.linspace(interval[0], interval[1], n)
    elif method == 'chebyshev':
        return np.cos((2 * np.arange(1, n + 1) - 1) / (2 * n) * np.pi) * (interval[1] - interval[0]) / 2 + (interval[1] + interval[0]) / 2
    else:
        raise ValueError("Método no reconocido: use 'equiespaciados' o 'chebyshev'.")

# Interpolación de Newton
def newton_divided_diff(x, y):
    n = len(x)
    coef = np.copy(y)
    for j in range(1, n):
        coef[j:n] = (coef[j:n] - coef[j - 1]) / (x[j:n] - x[j - 1])
    def newton_polynomial(x_val):
        result = coef[-1]
        for i in range(n - 2, -1, -1):
            result = result * (x_val - x[i]) + coef[i]
        return result
    return newton_polynomial, coef

# Conversión de coeficientes de Newton a polinomio simbólico
def newton_to_symbolic(x, coef):
    n = len(coef)
    x_symbol = sp.Symbol('x')
    polynomial = coef[-1]
    for i in range(n - 2, -1, -1):
        polynomial = polynomial * (x_symbol - x[i]) + coef[i]
    return sp.expand(polynomial)

# Conversión de interpolación de Lagrange a polinomio simbólico
def lagrange_to_symbolic(lagrange_poly):
    x_symbol = sp.Symbol('x')
    return sp.expand(lagrange_poly(x_symbol))

# Conversión de interpolación baricéntrica a polinomio simbólico
def barycentric_to_symbolic(x_vals, y_vals):
    coefficients = np.polyfit(x_vals, y_vals, deg=len(x_vals) - 1)
    x_symbol = sp.Symbol('x')
    polynomial = sum(coeff * x_symbol**i for i, coeff in enumerate(coefficients[::-1]))
    return sp.expand(polynomial)

# Calcular errores absolutos
def absolute_error(f, p, x_vals):
    return np.abs(f(x_vals) - p(x_vals))

# Definir parámetros
nodos = [11, 21]
functions = [f1, f2, f3]
function_labels = ["sin(x)", "exp(-20x^2)", "1/(1+25x^2)"]
metodos_nodos = ['equiespaciados', 'chebyshev']

# Bucle anidado
for n in nodos:
    for metodo in metodos_nodos:
        for func, func_label in zip(functions, function_labels):
            print(f"\nProcesando {func_label} con {n} nodos ({metodo})")

            # Generar nodos y valores
            x_vals = generate_nodes(n, method=metodo, func=func)
            y_vals = func(x_vals)
            x_plot = generate_nodes(5000, method='equiespaciados', func=func)
            y_true = func(x_plot)

            # Interpolación Baricéntrica
            start_time = time()
            y_bary = barycentric_interpolate(x_vals, y_vals, x_plot)
            bary_time = time() - start_time
            bary_error = absolute_error(func, lambda x: barycentric_interpolate(x_vals, y_vals, x), x_plot)
            bary_poly_sym = barycentric_to_symbolic(x_vals, y_vals)

            # Interpolación de Lagrange
            start_time = time()
            lagrange_poly = lagrange(x_vals, y_vals)
            lagrange_time = time() - start_time
            lagrange_poly_sym = lagrange_to_symbolic(lagrange_poly)
            lagrange_func = sp.lambdify(sp.Symbol('x'), lagrange_poly_sym)
            lagrange_error = absolute_error(func, lagrange_func, x_plot)

            # Interpolación de Newton
            start_time = time()
            newton_poly, newton_coef = newton_divided_diff(x_vals, y_vals)
            newton_time = time() - start_time
            newton_poly_sym = newton_to_symbolic(x_vals, newton_coef)
            newton_func = sp.lambdify(sp.Symbol('x'), newton_poly_sym)
            newton_error = absolute_error(func, newton_func, x_plot)
            print(f"Error medio Newton: {newton_error}")
            print(f"Error medio Baricéntrico: {bary_error}")
            print(f"Error medio Lagrange: {lagrange_error}")

            # Imprimir resultados
            print(f"Polinomio método Baricéntrico: {bary_poly_sym}")
            print(f"Tiempo Baricéntrico: {bary_time:.8f} s")
            print(f"Error medio Baricéntrico: {np.mean(bary_error):.8e}")
            print(f"Polinomio Lagrange: {lagrange_poly_sym}")
            print(f"Tiempo Lagrange: {lagrange_time:.8f} s")
            print(f"Error medio Lagrange: {np.mean(lagrange_error):.8e}")
            print(f"Polinomio Newton: {newton_poly_sym}")
            print(f"Tiempo Newton: {newton_time:.8f} s")
            print(f"Error medio Newton: {np.mean(newton_error):.8e}")

            # Graficar cada polinomio junto con la función original
            plt.figure(figsize=(10, 6))
            plt.plot(x_plot, y_true, label=f"{func_label} (Real)", linewidth=2)
            plt.plot(x_plot, y_bary, '--', label="Interpolación Baricéntrica")
            plt.plot(x_plot, lagrange_func(x_plot), '--', label="Interpolación Lagrange")
            plt.plot(x_plot, newton_func(x_plot), '--', label="Interpolación Newton")
            plt.scatter(x_vals, y_vals, color='red', label="Nodos")
            plt.title(f"{func_label} con {n} nodos ({metodo})")
            plt.xlabel("x")
            plt.ylabel("f(x)")
            plt.legend()
            plt.grid()
            plt.show()

            # Graficar errores
            plt.figure(figsize=(10, 6))
            plt.plot(x_plot, bary_error, label="Error Baricéntrico")
            plt.plot(x_plot, lagrange_error, label="Error Lagrange")
            plt.plot(x_plot, newton_error, label="Error Newton")
            plt.title(f"Errores absolutos: {func_label}, {n} nodos ({metodo})")
            plt.xlabel("x")
            plt.ylabel("Error absoluto")
            plt.legend()
            plt.grid()
            plt.show()
```
